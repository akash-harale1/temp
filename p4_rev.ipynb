```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
f = lambda x : (x + 3)**2
f_prime = lambda x : 2 * (x + 3)
x, lr, tol = 2, 0.1, 1e-6
x_vals = [x]

while(abs(f_prime(x)) > tol):
    x -= lr * f_prime(x)
    x_vals.append(x)

plt.plot(np.linspace(-6, 4, 100), f(np.linspace(-6, 4, 100)))
plt.plot(x_vals, f(np.array(x_vals)), 'ro--')
plt.show()
```
